# Creating indicator variables for categorical columns using Binary Encoder

* Went with Binary Encoder over Ordinal Encoding here since my data does not necesarily comprised of ordered variables. Binary encoding keeps the dimensionality down (vs. One Hot Encoding) and saves on memory. 

```python
encoder = ce.BinaryEncoder(cols=['gender','ever_married','work_type','Residence_type','smoking_status'],return_df=True,drop_invariant=False)
encoded_df = encoder.fit_transform(df)
encoded_df.head()
```

```python
encoder = ce.BinaryEncoder(cols=['gender','ever_married','work_type','smoking_status'],return_df=True)
encoded_df_nores = encoder.fit_transform(df_nores)
encoded_df_nores.head()
```

# Logistic Regression Model

* The predicted outcome for this model needs to be binary -- stroke negative(0) or stroke positive(1) so I want to start looking into models that support that. First I'll try logistic regression where I'll be predicting the probability of a particular outcome rather than the outcome itself.  

```python
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score
```

```python
clf = LogisticRegression(class_weight='balanced')
```

* I used the class_weight argument to signal to the model that the dataset is not balanced. There are many more cases of patients being stroke negative than there are stroke positive in our dataset. 

```python
clf.fit(X_train,y_train)
```

```python
y_predlr = clf.predict(X_test)
```

```python
print('R-squared value:',clf.score(X_train,y_train))
```

```python
accscore = accuracy_score(y_predlr,y_test)
print('Accuracy score:',accscore)
```

* Accuracy score is not very high but for logistic regression this score is just giving the probability of getting a particular outcome and not the accuracy of the model? 

```python
roc_score = roc_auc_score(y_test,y_predlr)
roc_score
```

```python
f1_score(y_test,y_predlr)
```

```python
from sklearn.metrics import confusion_matrix

tn,fp,fn,tp = confusion_matrix(y_test,y_predlr).ravel()
print(tn,fp,fn,tp)
```

```python
target_names = [0,1]
cm = confusion_matrix(y_test,y_predlr)

def plot_confusion_matrix(cm,classes,title='Confusion matrix',cmap=plt.cm.Blues):
    print(cm)
    
    plt.imshow(cm, interpolation = 'nearest',cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks,classes,rotation=45)
    plt.yticks(tick_marks,classes)
    
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.tight_layout
    
plt.figure()
plot_confusion_matrix(cm,classes=target_names,title='Confusion Matrix')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

clf = LogisticRegression(class_weight='balanced')
Cs = {'C':[0.001,0.1,1,10,100]}
model = GridSearchCV(clf,Cs,cv=5,scoring='accuracy')
model.fit(X_train,y_train)
model.best_estimator_,model.best_score_
```

I tried using the parameters suggested by GridSearchCV but i was getting worse results than when I just used the default C=1 so I left the 'C' argument as the default(C=1) in the orginal model.

In the case of diagnosing a patients stroke risk, it seems more valuable to have fewer false negatives than to worry about having more false positives.

# Summary

* Out of the three types of Naive Bayes models, the Gaussian Naive Bayes model seems to perform the best. Bernoulli and Multinomial produce better accuracy scores but the f1 score and roc_auc_score are lacking in a big way. The roc_auc_score for both are no better/barely better than if the data were pulled at random. I originally expected the Bernoulli model to work the best but it incorrectly identified all of the patients who were stroke positive resulting in high Type II error. 

* While the Gaussian Naive Bayes and Random Forest models performs better than the Logistic Regression model in terms of score, the confusion matrix still shows that there is a greater amount of false negatives. In the case of diagnosing a patients stroke risk, it seems right to go with the model that predicts fewer false negative results. 

* Naive Bayes does also make the assumption that all features are independent of each other; in our case, I don't think it's safe to say our features are entirely independent of each other. For example, a patient with a high bmi could be more likely to have hypertension, heart disease, or diabetes.
